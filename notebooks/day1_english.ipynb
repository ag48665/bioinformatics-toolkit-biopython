# Biopython Day 1: Sequence Basics

This notebook introduces the core Biopython objects used in the project.

## 1. Creating a DNA sequence with `Seq`
`Seq` is the basic object for DNA, RNA, and protein sequences.

```python
from Bio.Seq import Seq

dna = Seq("ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG")
print(dna)
print("Length:", len(dna))
```

## 2. Complement and reverse complement
The complement replaces each DNA base with its paired base. The reverse complement also reverses the order.

```python
print("Complement:", dna.complement())
print("Reverse complement:", dna.reverse_complement())
```

## 3. Transcription and translation
Transcription converts DNA to RNA. Translation converts RNA or DNA codons to amino acids.

```python
rna = dna.transcribe()
protein = rna.translate()
print("RNA:", rna)
print("Protein:", protein)
```

## 4. GC content
GC content is the percentage of G and C bases in a DNA sequence.

```python
from Bio.SeqUtils import gc_fraction

gc_percent = gc_fraction(dna) * 100
print(f"GC content: {gc_percent:.2f}%")
```

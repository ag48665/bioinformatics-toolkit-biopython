## 1. FASTA parsing with `SeqIO`
`SeqIO.parse()` reads sequence files and returns `SeqRecord` objects.

```python
from Bio import SeqIO

records = list(SeqIO.parse("../data/example.fasta", "fasta"))
for record in records:
    print(record.id, record.description, len(record.seq))
```

## 2. `SeqRecord` fields
A `SeqRecord` contains the sequence plus metadata.

```python
record = records[0]
print("ID:", record.id)
print("Name:", record.name)
print("Description:", record.description)
print("Sequence:", record.seq)
print("Annotations:", record.annotations)
```

## 3. Protein sequence analysis
A protein sequence can be counted with normal Python tools.

```python
from collections import Counter
protein = "MAIVMGRWKGAR"
print("Length:", len(protein))
print(Counter(protein))
```

## 4. Pairwise alignment
Pairwise alignment compares two biological sequences.

```python
from Bio import pairwise2
from Bio.pairwise2 import format_alignment

alignments = pairwise2.align.globalxx("ACCGT", "ACG")
print(format_alignment(*alignments[0]))
```

## 5. Motifs and consensus sequences
A motif represents a recurring pattern in DNA, RNA, or protein sequences.

```python
from Bio import motifs
from Bio.Seq import Seq

instances = [Seq("ATGC"), Seq("ATGA"), Seq("ATGT")]
motif = motifs.create(instances)
print("Consensus:", motif.consensus)
print("Degenerate consensus:", motif.degenerate_consensus)
```

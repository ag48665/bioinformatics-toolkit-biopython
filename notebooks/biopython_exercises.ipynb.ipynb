```python

from Bio.Seq import Seq

dna = Seq("ATGGCCATTGTAATGGGCCGCTGAAAGGGTGCCCGATAG")
print(dna)
print("Length:", len(dna))
     
```

```python

print("Complement:", dna.complement())
print("Reverse complement:", dna.reverse_complement())
```

```python

rna = dna.transcribe()
protein = rna.translate()
print("RNA:", rna)
print("Protein:", protein)
```

```python

from Bio.SeqUtils import gc_fraction

gc_percent = gc_fraction(dna) * 100
print(f"GC content: {gc_percent:.2f}%")
```

```python
from Bio import SeqIO
records = list(SeqIO.parse("example.fasta", "fasta"))
for record in records:
    print(record.id, record.description, len(record.seq))
```

```python
record = records[0]
print("ID:", record.id)
print("Name:", record.name)
print("Description:", record.description)
print("Sequence:", record.seq)
print("Annotations:", record.annotations)  
```

```python
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
alignments = pairwise2.align.globalxx("ACCGT", "ACG")
print(format_alignment(*alignments[0]))
```

```python
from Bio import motifs
from Bio.Seq import Seq
instances = [Seq("ATGC"), Seq("ATGA"), Seq("ATGT")]
motif = motifs.create(instances)
print("Consensus:", motif.consensus)
print("Degenerate consensus:", motif.degenerate_consensus)
```

```python
from Bio.PDB import PDBParser
parser = PDBParser(QUIET=True)
structure = parser.get_structure(
    "hemoglobin",
    "data/1A3N.pdb"
)
print("Structure ID:", structure.id)
print("Number of chains:", len(list(structure.get_chains())))
print("Number of residues:", len(list(structure.get_residues())))
print("Number of atoms:", len(list(structure.get_atoms())))
```

```python
from Bio import Entrez

# Entrez provides access to NCBI databases such as PubMed,
# GenBank, Protein and Gene. It allows users to search,
# retrieve and analyze biological records programmatically.

# Example: search PubMed for articles related to Biopython

Entrez.email = "your.email@example.com"

handle = Entrez.esearch(
    db="pubmed",
    term="biopython",
    retmax=5
)

record = Entrez.read(handle)
handle.close()

print("PubMed IDs:")
for pmid in record["IdList"]:
    print(pmid)

# Common Entrez databases:
# - pubmed      : scientific publications
# - nucleotide  : DNA/RNA sequences
# - protein     : protein sequences
# - gene        : gene information
# - structure   : 3D molecular structures
```
